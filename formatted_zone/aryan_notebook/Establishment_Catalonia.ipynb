```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/raw/establishments_catalonia.csv')
df.head()
```

```python
df = df[['id', 'commercial_name', 'county_code', 'full_address', 'UTMx', 'UTMy', 'location']]
df['latitude'] = df['location'].str.split(',').str[0].astype(float)
df['longitude'] = df['location'].str.split(',').str[1].astype(float)
df.drop(columns=['location'], inplace=True)
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.head()
```

```python
df['county_code'].value_counts()
```

```python
df.to_csv('../../data/aryan_pandas/establishment_catalonia.csv', index=False)
```
