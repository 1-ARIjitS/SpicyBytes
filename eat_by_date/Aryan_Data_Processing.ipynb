```python
# find dictionary keys which have item_description as ''

items_to_delete = []  # those with item_description as ''

items_without_description = []

for dict_item in data:
    try:
        if dict_item['item_description'] == '':
            items_to_delete.append(dict_item)
    except:
        items_without_description.append(dict_item)
#         print(dict_item)
```

```python
# find dictionary keys which have item_description as ''

items_to_delete = []  # those with pantry or refrigerator as 'Past Printed Date'

items_without_description = []  # those that do not have pantry or refrigerator keys

for dict_item in data:
    try: 
        if 'Past Printed Date' in dict_item.values():
            items_to_delete.append(dict_item)
        elif 'Past Date' in dict_item.values():
            items_to_delete.append(dict_item)
    except:
        items_without_description.append(dict_item)
```
